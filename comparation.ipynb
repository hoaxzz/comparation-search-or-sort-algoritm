```python
#pertama kita harus menginisialisasi
df = pd.read_csv(path)

data_amount = df['amount'].tolist()  # buat list numerik
data_random = df['transaction_id'].tolist()  # buat list string untuk sentinel search
```

```python
# ================================================================
# 2. SEARCH ALGORITHMS
# ================================================================

# ---------- Sentinel Linear Search ----------
def sentinel_linear_search(arr, key):
    arr.append(key)  # tambahkan sentinel
    i = 0
    while arr[i] != key:
        i += 1
    arr.pop()  # hapus sentinel
    return i if i < len(arr) else -1

# ---------- Interpolation Search ----------
def interpolation_search(arr, key):
    low = 0
    high = len(arr) - 1

    while low <= high and key >= arr[low] and key <= arr[high]:
        if low == high:
            return low if arr[low] == key else -1

        # Rumus interpolasi
        pos = low + int(((float(high - low) /
                         (arr[high] - arr[low])) * (key - arr[low])))

        if arr[pos] == key:
            return pos
        if arr[pos] < key:
            low = pos + 1
        else:
            high = pos - 1
    return -1
```

```python
# ================================================================
# 3. SORTING ALGORITHMS
# ================================================================

# ---------- Gnome Sort ----------
def gnome_sort(arr):
    i = 0
    while i < len(arr):
        if i == 0 or arr[i] >= arr[i - 1]:
            i += 1
        else:
            arr[i], arr[i-1] = arr[i-1], arr[i]
            i -= 1
    return arr

# ---------- Comb Sort ----------
def comb_sort(arr):
    n = len(arr)
    gap = n
    shrink = 1.3
    sorted_flag = False

    while not sorted_flag:
        gap = int(gap / shrink)
        if gap <= 1:
            gap = 1
            sorted_flag = True

        i = 0
        while i + gap < n:
            if arr[i] > arr[i + gap]:
                arr[i], arr[i + gap] = arr[i + gap], arr[i]
                sorted_flag = False
            i += 1
    return arr
```

```python
# ================================================================
# 4. EKSEKUSI DAN PERBANDINGAN WAKTU
# ================================================================

# -------------------- SEARCH TEST --------------------
import random
from time import time

key_search_amount = random.choice(data_amount)
key_search_id = random.choice(data_random)

# Sentinel Linear Search
arr1 = data_random.copy()
t1 = time()
sentinel_linear_search(arr1, key_search_id)
t1 = time() - t1

# Interpolation Search (sorted)
arr2 = sorted(data_amount.copy())
t2 = time()
interpolation_search(arr2, key_search_amount)
t2 = time() - t2

# -------------------- SORT TEST --------------------
arr3 = data_amount.copy()
t3 = time()
gnome_sort(arr3)
t3 = time() - t3

arr4 = data_amount.copy()
t4 = time()
comb_sort(arr4)
t4 = time() - t4
```

```python
# ================================================================
# 5. OUTPUT HASIL
# ================================================================
print("\n=== HASIL KOMPARASI SEARCH ===")
print(f"Sentinel Linear Search : {t1:.6f} detik")
print(f"Interpolation Search    : {t2:.6f} detik")

print("\n=== HASIL KOMPARASI SORT ===")
print(f"Gnome Sort : {t3:.6f} detik")
print(f"Comb Sort  : {t4:.6f} detik")
```
